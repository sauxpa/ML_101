```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import CubicSpline
from matplotlib.ticker import FuncFormatter
```

```python
measure_times = [0, 3, 12, 24, 60]
```

### Interpolate in weights

```python
measured_weights = [150, 140, 105, 90, 100]

spline_nak = CubicSpline(measure_times, measured_weights, bc_type='not-a-knot')
spline_cl = CubicSpline(measure_times, measured_weights, bc_type='clamped')
spline_nat = CubicSpline(measure_times, measured_weights, bc_type='natural')

tt = np.linspace(0, measure_times[-1] + 1, 500)

with sns.axes_style('darkgrid'):
    fig, ax = plt.subplots(figsize=(7, 5), nrows=1, ncols=1)
    ax.plot(tt, spline_nak(tt), color='red', alpha=0.6, linestyle='dashed', label='Not-a-knot')
    ax.plot(tt, spline_cl(tt), color='orange', alpha=0.6, linestyle='dashed', label='Clamped')
    ax.plot(tt, spline_nat(tt), color='green', alpha=0.9, label='Natural')
    ax.scatter(measure_times, measured_weights, marker='s', label='measured')
    
    ax.legend()
    ax.set_xlabel('Number of months')
    ax.set_ylabel('Weight (kg)')
    ax.set_title('Interpolated weight curve')
    plt.tight_layout()
    plt.show()
```
